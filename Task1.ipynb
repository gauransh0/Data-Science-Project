```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df1=pd.read_csv(r"C:\Users\HP\Downloads\data\data\applicant.csv")
df2=pd.read_csv(r"C:\Users\HP\Downloads\data\data\loan.csv")
```

```python
df=pd.merge(df1,df2)  #merging the 2 dataset for clear understanding of data
df.head()             #looking at first 5 rows of merged dataset
```

```python
#checking for duplicate values
df.duplicated().sum()
```

```python
# count of null values
df.isnull().sum()
```

```python
df['Purpose'].value_counts().plot(kind='barh')
#plotting the bar graph horizontally for purpose
```

```python
#replacing the null value with the mode as it's categorical
df['Purpose'].fillna(df['Purpose'].mode()[0],inplace=True) 
df['Property'].value_counts().plot(kind='barh')
```

```python
df['Has_been_employed_for_at_least'].value_counts().plot(kind='barh')
```

```python
df['Has_been_employed_for_at_least'].fillna(df['Has_been_employed_for_at_least'].mode()[0],inplace=True)
df['Has_been_employed_for_at_least'].value_counts().plot(kind='barh')
```

```python
#doing the same steps for all categorical data .
df['Has_been_employed_for_at_most'].fillna(df['Has_been_employed_for_at_most'].mode()[0],inplace=True)
df['Savings_account_balance'].fillna(df['Savings_account_balance'].mode()[0],inplace=True)
df['Property'].fillna(df['Property'].mode()[0],inplace=True)
df.isnull().sum()
```

```python
#removing all null value coloumn 
df.drop(columns=['Telephone'],inplace=True)
df.drop(columns=['Balance_in_existing_bank_account_(lower_limit_of_bucket)'],inplace=True)
df.drop(columns=['Balance_in_existing_bank_account_(upper_limit_of_bucket)'],inplace=True)
df.drop(columns=['Other_EMI_plans'],inplace=True)
df.isnull().sum() #checking null again
```

```python
#data set with no null values
df.head()
```

```python
# segmenting customers based on high risk
df.corr()["high_risk_applicant"]
```

```python
# removing the negative co relation with the data
df.drop(columns=['applicant_id'],inplace=True)
df.drop(columns=['Has_guarantor'],inplace=True)
df.drop(columns=['Number_of_existing_loans_at_this_bank'],inplace=True)
df.drop(columns=['Primary_applicant_age_in_years'],inplace=True)
df.drop(columns=['Number_of_dependents'],inplace=True)
df.corr()["high_risk_applicant"]
```

```python
df.head() #checking data again
```

```python
df['log_Principal_loan_amount']=np.log(df['Principal_loan_amount']) #normalising right skew of 'Principal_loan_amount'
df['log_Principal_loan_amount'].hist(bins=20)
```

```python
df['Months_loan_taken_for'].hist(bins=30)
```
